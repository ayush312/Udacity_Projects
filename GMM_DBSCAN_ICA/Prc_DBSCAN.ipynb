```python
import numpy as np
import pandas as pd
import dbscan_lab_helper as helper

%matplotlib inline

data = pd.read_csv('blobs.csv')
data_1 = data.values[:80]
helper.plot_dataset(data_1)
```

```python
# lets do DBSCAN on this dataset
from sklearn.cluster import DBSCAN
pred_1 = DBSCAN().fit_predict(data_1)

helper.plot_clustered_dataset(data_1, pred_1)
```

```python
helper.plot_clustered_dataset(data_1, pred_1, neighborhood=True)
```

```python
pred_2 = DBSCAN(eps=2).fit_predict(data_1)

helper.plot_clustered_dataset(data_1, pred_2, neighborhood=True, epsilon = 2)
```

```python
# lets have new dataset
data_2 = pd.read_csv('varied.csv')
data_2 = data_2.values[:300]
helper.plot_dataset(data_2, xlim=(-14, 5), ylim=(-12, 7))
```

```python
# do default DBSCAN
pred_3 = DBSCAN().fit_predict(data_2)

helper.plot_clustered_dataset(data_2, 
                              pred_3, 
                              xlim=(-14, 5), 
                              ylim=(-12, 7), 
                              neighborhood=True, 
                              epsilon=0.5)
```

```python
eps = 1.5
min_sample= 50

pred_4 = DBSCAN(eps=eps, min_samples=min_sample).fit_predict(data_2)


helper.plot_clustered_dataset(data_2, 
                              pred_4, 
                              xlim=(-14, 5), 
                              ylim=(-12, 7), 
                              neighborhood=True, 
                              epsilon=eps)
```

```python
eps = [0.4, 0.6, 1, 1.5, 2]
min_sample= [2,5,10,25,50]

helper.plot_dbscan_grid(data_2,eps,min_sample)
```
